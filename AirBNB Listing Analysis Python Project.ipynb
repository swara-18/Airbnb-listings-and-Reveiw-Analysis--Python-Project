```python
Filter the data down to rows where the city is Paris, and 
keep only the columns ‘host_since’, ‘neighbourhood’, 
‘city’, ‘accommodates’, and ‘price’
```

```python
paris_listings= (
    listings
    .query("city=='Paris'")
    .loc[:,["host_since", "neighbourhood", "city", "accommodates", "price"]])
paris_listings.info()
```

```python
Create a table named paris_listings_neighbourhood that 
groups Paris listings by 'neighbourhood' and calculates 
the mean price (sorted low to high)
```

```python
paris_listings_neighbourhood= (
    paris_listings
    .groupby('neighbourhood')
    .agg({"price": "mean" })
    .sort_values("price")
)
paris_listings_neighbourhood.head()
```

```python
Create a table named paris_listings_accomodations, 
filter down to the most expensive neighborhood, 
group by the ‘accommodations’ column, 
and add the mean price for each value of ‘accommodates’ (sorted low to high)
```

```python
paris_listings_accomodations=  (
    paris_listings
    .query("neighbourhood == 'Elysee'")
    .groupby('accommodates')
    .agg({"price":"mean", "neighbourhood":"count"})
    .sort_values("accommodates")
)
paris_listings_accomodations.head()
```

```python
Create a table called paris_listings_over_time 
grouped by the ‘host_since’ year, and calculate 
the average price and count of rows representing the number of new hosts

```

```python
paris_listings_over_time = (
    paris_listings
    .set_index("host_since")
    .resample("y")
    .agg({"neighbourhood": "count", 
        "price":"mean" 
         })
)
paris_listings_over_time.head()
```

```python
Create a horizontal bar chart of the average price by 
neighborhood in Paris, and make sure to add a title
and change axis labels as needed
```

```python
import matplotlib.pyplot as plt
(paris_listings_neighbourhood
 .plot
 .barh(
     title="Average Price by Neighbourhood in Paris",
     xlabel= "price",
     ylabel= "Neighbourhood",
     legend= None,
     color='g')
)
plt.show()
```

```python
Create two line charts: one showing the count of new hosts over time, 
and one showing average price. 
Set the y-axis limit to 0, add a title, and change axis labels as needed
```

```python
paris_listings_over_time["neighbourhood"].plot(
    ylabel="New Hosts",
    title="New AirBnb Hosts in paris over time"
)
plt.show()
paris_listings_over_time["price"].plot(
    xlabel="Average Price",
    title="AirBnb's Average Price in paris"
)
plt.show()
    
```
